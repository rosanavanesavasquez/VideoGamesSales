```python
#Importo todas las librerias
import pandas as pd
from datetime import datetime, timedelta
from matplotlib import pyplot as plt
from matplotlib import dates as mpl_dates
import seaborn as sns
from mpl_toolkits.mplot3d import Axes3D

# Mi Carpeta VideoGamesSales
dataset_path = '/content/gdrive/MyDrive/CODER HOUSE/DATA_SET/VideoGamesSales/Video_Games_Sales_as_at_22_Dec_2016.csv'

# DataFrame
vg_sales = pd.read_csv(dataset_path)
```

```python
# Ver la cantidad de valores faltantes en cada columna
print(vg_sales.isna().sum())
```

```python
plt.figure(figsize=(10, 7))
sns.scatterplot(x='Year_of_Release', y='Global_Sales', hue='Genre', data=vg_sales)

plt.xlabel('Year of Release')
plt.ylabel('Global Sales')
plt.title('Sales by Year and Genre')
plt.show()
```

Gráfico de barras por género y ventas globales:

```python
vg_sales_grouped = vg_sales.groupby('Genre')['Global_Sales'].sum().sort_values(ascending=False)

plt.figure(figsize=(10, 6))
vg_sales_grouped.plot(kind='bar', color='skyblue')
plt.title('Total Global Sales by Genre')
plt.xlabel('Genre')
plt.ylabel('Global Sales (in millions)')
plt.xticks(rotation=45)
plt.show()
```

Gráfico de líneas de ventas globales por año:

```python
sales_by_year = vg_sales.groupby('Year_of_Release')['Global_Sales'].sum()

plt.figure(figsize=(10, 6))
plt.plot(sales_by_year.index, sales_by_year.values, marker='o', linestyle='-', color='green')
plt.title('Global Sales Over Time')
plt.xlabel('Year of Release')
plt.ylabel('Global Sales (in millions)')
plt.grid(True)
plt.show()
```

```python
#Elimino valores nulos
vg_sales_cleaned = vg_sales.dropna(subset=['Critic_Score', 'User_Score'])
```

```python
vg_sales['User_Score'].fillna(vg_sales['User_Score'].median(), inplace=True)
```

```python
plt.figure(figsize=(10, 6))
vg_sales.groupby('Year_of_Release')[['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']].sum().plot(kind='line')
plt.title('Regional Sales Over Time')
plt.xlabel('Year of Release')
plt.ylabel('Sales (in millions)')
plt.grid(True)
plt.legend(['North America', 'Europe', 'Japan', 'Other'])
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Genre', y='Global_Sales', data=vg_sales)
plt.xticks(rotation=45)
plt.title('Distribution of Global Sales by Genre')
plt.ylabel('Global Sales (in millions)')
plt.show()
```

```python
plt.figure(figsize=(10, 7))
sns.scatterplot(x='Critic_Score', y='User_Score', data=vg_sales)
plt.title('Critic Score vs User Score')
plt.xlabel('Critic Score')
plt.ylabel('User Score')
plt.show()
```
